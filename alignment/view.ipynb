```python
import matplotlib.pyplot as plt
import uproot
import hist
from _differ import Differ
import mplhep
plt.style.use(mplhep.style.ROOT)

def diff(title, pass_name) :
    return Differ(title,
           (f'kf/HPS_Nominal_2019SensorSurvey_iter0/{pass_name}_gblplots.root',
            'KF Nominal', dict(ls='solid', color='tab:blue')),
           (f'kf/HPS_2019_L1ty100um_iter0/{pass_name}_gblplots.root',
            'KF Misalign', dict(ls='solid', color='tab:orange')),
           (f'st/HPS_Nominal_2019SensorSurvey_iter0/{pass_name}_gblplots.root',
            'ST Nominal', dict(ls='dashed', color='tab:blue')),
           (f'st/HPS_2019_L1ty100um_iter0/{pass_name}_gblplots.root',
            'ST Misalign', dict(ls='dashed', color='tab:orange'))
          )
```

```python
d = diff('L1 Top $\Delta y = +100 \mu$m', 'debug')
```

```python
d.plot1d('trk_params/nHits_top','N Hits Top',ylabel='Count',yscale='linear',legend_kw=dict(loc='upper left'))
d.plot1d('trk_params/nHits_bottom', 'N Hits Bottom', ylabel='Count', yscale='linear', legend_kw=dict(loc='upper left'))
```

```python
sensors = [
    ('L1b Axial', 'res/uresidual_GBL_module_L1b_halfmodule_axial_sensor0'),
    ('L1b Stereo', 'res/uresidual_GBL_module_L1b_halfmodule_stereo_sensor0'),
    ('L1t Axial', 'res/uresidual_GBL_module_L1t_halfmodule_axial_sensor0'),
    ('L1t Stereo', 'res/uresidual_GBL_module_L1t_halfmodule_stereo_sensor0'),
]
for name, hist in sensors :
    d.plot1d(hist, f'Unbiased Residual {name} [mm]', ylabel='Count')
```
