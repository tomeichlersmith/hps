# View Track Histograms

```python
import matplotlib.pyplot as plt
import uproot
import hist
import mplhep

def comp(gbl, kf, hist_name) :
    ax = plt.figure(1).subplots()
    gbl[hist_name].to_hist().plot(ax=ax,label='GBL')
    kf[hist_name].to_hist().plot(ax=ax,label='KF')
    ax.legend()
    ax.set_xlabel(hist_name)
    plt.show()
```

```python
f = uproot.open('mc_events_gbl_gblplots.root')
```

```python
kf = uproot.open('mc_events_kf_kfgblplots.root')
```

```python
for h in ['Chi2_bottom','Chi2_bottom_neg','Chi2_bottom_pos','Chi2_top','Chi2_top_neg','Chi2_top_pos'] :
    comp(f,kf,f'trk_params/{h}')
```
