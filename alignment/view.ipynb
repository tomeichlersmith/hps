# View Track Histograms

```python
import matplotlib.pyplot as plt
import uproot
import hist
import mplhep
plt.style.use(mplhep.style.ROOT)
```

```python
f = uproot.open('mc_events_gbl_gblplots.root')
```

```python
kf = uproot.open('mc_events_kf_kfgblplots.root')
```

```python
def comp(gbl, kf, hist_name, hist_kw=dict()) :
    ax = plt.figure(1, figsize=(11,8)).subplots()
    gbl[hist_name].to_hist().plot(ax=ax,label='GBL',**hist_kw)
    kf[hist_name].to_hist().plot(ax=ax,label='KF',**hist_kw)
    ax.legend()
    ax.set_xlabel(hist_name)
    ax.set_yscale('log')
    plt.show()
```

```python
for h in ['Chi2_bottom','Chi2_bottom_neg','Chi2_bottom_pos','Chi2_top','Chi2_top_neg','Chi2_top_pos'] :
    comp(f,kf,f'trk_params/{h}')
```

```python
for h in ['uresidual_GBL_top'] :
    comp(f,kf,f'res/{h}')
```

```python
comp(f,kf,'trk_params/nHits_top')
```
